# Completed tasks


The number of tasks performed according to the level of well-behaved participants

```python
import random as rand


def calculateRequesterBalance(decisionProbability, taskCount):
  if decisionProbability < 0 or decisionProbability > 1:
    raise ValueError(decisionProbability, "Args must be between 0 and 1")
  else:
    result = 0
    for i in range(0, taskCount):

      # Requester waits until task assign
      if rand.random() / 2 <= decisionProbability:

        # Requester waits until task verify
        if rand.random() <= decisionProbability:
          
          # Requester finishes the task
          if rand.random() <= decisionProbability:

            result = result + 1

        # Reassign task

    return result / taskCount

  return


```

```python
import matplotlib.pyplot as plt
import numpy as np


def showFinishedTasksCount(testCount):
  probabilities = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]
  outputs = [calculateRequesterBalance(i, testCount) * testCount for i in probabilities]
  plt.plot(probabilities, outputs, 'b-')
  plt.xlabel("good behavior probability")
  plt.ylabel("number of finished tasks")
  plt.show()

showFinishedTasksCount(100)
```
