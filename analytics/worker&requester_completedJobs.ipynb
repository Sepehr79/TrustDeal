### Number of Completed Task

In this repository we will see how good behavior effects the count of completed tasks for both requester and worker.

```python
import random

def calculateWorkerCompletedTasks(decisionProbability, rangeNumber):
  if decisionProbability < 0 or decisionProbability > 1:
    raise ValueError(decisionProbability, "Args must be between 0 and 1")
  else:
    counter = 0
    for i in range(0, rangeNumber):
      # Verify Task
      if random.random() <= decisionProbability:
        # Wait until finish task
        if random.random() <= decisionProbability:
          counter = counter + 1
        # Reassign
        
    return (counter / rangeNumber)
  
def calculateRequesterCompletedTasks(decisionProbability, testCounter):
  if decisionProbability < 0 or decisionProbability > 1:
    raise ValueError(decisionProbability, "Args must be between 0 and 1")
  else:

    counter = 0
    for i in range(0, testCounter):

      # Requester waits until assign task
      if  random.random() / 100 <= decisionProbability:

        # Requester waits until verify task
        if  random.random() * 95 / 100 <= decisionProbability:

          # Requester finished the task
          if  random.random() * 95 / 100 <= decisionProbability:
          
            counter = counter + 1

  return (counter / testCounter)


import matplotlib.pyplot as plt


def showFinishedTasksCount(testCount):
  probabilities = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]
  outputs_requester = [calculateRequesterCompletedTasks(i, testCount) for i in probabilities]
  outputs_worker = [calculateWorkerCompletedTasks(i, testCount) for i in probabilities]
  plt.plot(probabilities, outputs_requester, 'r-', label="requester")
  plt.plot(probabilities, outputs_worker, 'y-', label="worker")
  plt.xlabel("good behavior probability")
  plt.ylabel("number of finished tasks")
  plt.show()

showFinishedTasksCount(500)

```
